## Problem 1.
<br>
<span style="font-size:120%">
    Estimate the $\mathbf{VAR}(4)$ model for the growth rates of $\mathbf{GDP}$, capital formation, private consumption and export using the excel data uploaded in the class webpage.

```python
data = pd.read_excel('GDP.xlsx', index_col = 0)
```

```python
data
```

```python
private_consumption = data.iloc[1,:]
private_consumption
```

```python
capital_formation = data.iloc[5,:]
capital_formation
```

```python
export = data.iloc[-8,:]
export
```

```python
gdp = data.iloc[-1,:]
gdp
```

```python
matrix = pd.concat([gdp, capital_formation, private_consumption, export], axis = 1)
matrix = pd.DataFrame(matrix.values, index = np.arange(1,78), columns = ['GDP','capital_formation','private_consumption','export']).astype('float')
matrix
```

```python
import seaborn as sns
sns.set_theme(style = "whitegrid", palette = "pastel")
plt.figure(figsize = (15,6))
sns.lineplot(data = matrix, markers = True, dashes = False)
plt.legend(fontsize = 15)
plt.xlabel('Period', fontsize = 17, fontweight = 'bold', fontfamily = 'Serif')
plt.ylabel('Index', fontsize = 17, fontweight = 'bold', fontfamily = 'Serif')
plt.title('Korean GDP (Original Data)', fontsize = 20, fontweight = "bold", fontfamily = 'Serif')
```

```python
matrix_period = matrix.pct_change(periods = 4).dropna()
```

```python
matrix_period
```

```python
plt.figure(figsize = (15,6))
sns.lineplot(data = matrix_period, markers = True, dashes = False)
plt.legend(fontsize = 15)
plt.xlabel('Period', fontsize = 17, fontweight = 'bold', fontfamily = 'Serif')
plt.ylabel('Index', fontsize = 17, fontweight = 'bold', fontfamily = 'Serif')
plt.title('Korean GDP (Seasonal Differencing)', fontsize = 20, fontweight = "bold", fontfamily = 'Serif')
```

```python
import statsmodels.api as sm
from statsmodels.tsa.api import VAR

model = VAR(matrix_period).fit(4, trend = 'c')
```

```python
model.summary()
```

```python
model.resid
```

```python
order = VAR(matrix_period).select_order(maxlags=4)
print(order)
```

## Problem 2.
<br>
<span style="font-size:120%">
    Does capital formation Granger-cause $\mathbf{GDP}$?

```python
gctest = model.test_causality('GDP', 'capital_formation', kind = 'wald')
gctest.summary()
```

<span style="font-size:120%">
    It appeared that Capital formation does not Granger-cause $\mathbf{GDP}$ on Korean data
